```python
# Import basic libraries
import pandas as pd
import numpy as np
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Import model libraries
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Read in datafile
df = pd.read_csv("data\\teleco_market_basket.csv")
pd.set_option('display.max_columns', None)
df.head(15)
```

```python
# Convert file data to data frame
df.dropna(subset = ["Item01"], inplace=True)
df = df.T.apply(lambda x: x.dropna().tolist()).tolist()
```

```python
# Use one hot encoding to create True/False value matrix for transactions
encoder = TransactionEncoder()
onehot = encoder.fit(df).transform(df)
df_apriori = pd.DataFrame(onehot, columns=encoder.columns_)
df_apriori.head(5)
```

```python
# Save encoded data to new .csv
df_apriori.to_csv("data\\teleco_market_basket_cleaned.csv")
```

```python
# Initialize apriori algorithm
apriori = mlxtend.frequent_patterns.apriori(df_apriori)
```

```python
# Calculating support
apriori_support = mlxtend.frequent_patterns.apriori(df_apriori, min_support=0.01, use_colnames=True)
apriori_support.head(20)
```

```python
# Printing rules
rules = association_rules(apriori_support, metric="support", min_threshold=0.01, support_only=False)
rules.head(50)
```

```python
# Top rules by confidence
rules2 = association_rules(apriori_support, metric="confidence", min_threshold=0.45, support_only=False)
rules2.head(20)
```

```python
# Top rules by support
rules3 = association_rules(apriori_support, metric="support", min_threshold=0.05, support_only=False)
rules3.head(20)
```

```python
# Top rules by lift
rules3 = association_rules(apriori_support, metric="lift", min_threshold=2.5, support_only=False)
rules3.head(20)
```

```python
# Top 3 rules by confidence
rules4 = association_rules(apriori_support, metric="confidence", min_threshold=0.47, support_only=False)
rules4.head(20)
```
